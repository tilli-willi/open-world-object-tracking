```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from PIL import Image


heatmap = np.load("/home/uig93971/src/data/street_obstacle_sequences/tracking_res/track_thresh0.1_dti/heatmaps/sequence_015/000032_raw_data.npy")
gt = np.array(Image.open("/home/uig93971/src/data/street_obstacle_sequences/semantic_ood/sequence_015/000032_semantic_ood.png"))
heatmap[gt == 255] = 0

# Define a custom colormap from green to red
cmap = mcolors.LinearSegmentedColormap.from_list("GreenRed", ["green", "red"])

# Visualize the heatmap
plt.imshow(heatmap, cmap=cmap)
plt.colorbar(label="Intensity (0: Green, 1: Red)")
plt.title("Heatmap Visualization")
plt.xlabel("Width")
plt.ylabel("Height")
plt.show()
```
